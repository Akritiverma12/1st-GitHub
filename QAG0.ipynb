```python
# --------------------------
# Install required packages
# --------------------------


# --------------------------
# Imports
# --------------------------
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
import nltk

from nltk.tokenize import sent_tokenize
from difflib import SequenceMatcher

# --------------------------
# Load Models
# --------------------------

# Question Generation model
qg_tokenizer = AutoTokenizer.from_pretrained("iarfmoose/t5-base-question-generator")
qg_model = AutoModelForSeq2SeqLM.from_pretrained("iarfmoose/t5-base-question-generator")

# Question Answering model
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")

# --------------------------
# QAGS Functions
# --------------------------

def generate_questions(summary):
    """Generate (question, answer) pairs from each sentence in the summary."""
    sentences = sent_tokenize(summary)
    qg_pairs = []
    for sentence in sentences:
        input_text = f"generate question: {sentence}"
        input_ids = qg_tokenizer.encode(input_text, return_tensors="pt")
        outputs = qg_model.generate(input_ids, max_length=64, num_return_sequences=1)
        question = qg_tokenizer.decode(outputs[0], skip_special_tokens=True)
        qg_pairs.append((question, sentence))  # question, expected answer
    return qg_pairs

def get_qa_answers(qg_pairs, source_text):
    """Answer generated questions using the source document."""
    results = []
    for question, expected_answer in qg_pairs:
        try:
            prediction = qa_pipeline({
                "context": source_text,
                "question": question
            })
            predicted_answer = prediction['answer']
        except:
            predicted_answer = ""
        results.append((question, expected_answer, predicted_answer))
    return results

def get_similarity(a, b):
    """Simple string similarity score."""
    return SequenceMatcher(None, a.lower(), b.lower()).ratio()

def qags_score(source_text, summary):
    """Main QAGS scoring function."""
    print("=== Generating Questions from Summary ===")
    qg_pairs = generate_questions(summary)

    print("\n=== Running QA on Source Text ===")
    qa_results = get_qa_answers(qg_pairs, source_text)

    scores = []
    for i, (question, expected, predicted) in enumerate(qa_results):
        sim = get_similarity(expected, predicted)
        scores.append(sim)
        print(f"\nExample {i+1}")
        print(f"Q: {question}")
        print(f"Expected: {expected}")
        print(f"Predicted: {predicted}")
        print(f"Similarity: {sim:.2f}")

    overall_score = sum(scores) / len(scores) if scores else 0.0
    print(f"\n✅ Final QAGS Score: {overall_score:.2f}")
    return overall_score

# --------------------------
# Example Usage
# --------------------------

source_doc = """
The Eiffel Tower is a famous iron structure located in Paris, France.
It was built in 1889 to serve as the entrance arch to the 1889 World's Fair.
Standing approximately 300 meters tall, it was the tallest structure in the world until 1930.
Today, it remains one of the most visited monuments globally.
"""

generated_summary = """
The Eiffel Tower was constructed in 1889 for the World's Fair in Paris.
It was the tallest building in the world at that time and remains a top tourist attraction.
"""

# Run QAGS score
qags_score(source_doc, generated_summary)
```
